## 3. Greatest among two numbers


```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest:", a)
elif b > a:
    print("Greatest:", b)
else:
    print("Numbers are equal")
```

## 4. Greatest among three numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Greatest:", a)
elif b > a and b > c:
    print("Greatest:", b)
else:
    print("Greatest:", c)
```


## 5. Quadratic Equation Roots



```python
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b*b - 4*a*c   # discriminant

if d >= 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Roots are:", r1, "and", r2)
else:
    print("Roots are imaginary")
```

## 6. Leap Year

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

## 7. Next Date Program

```python
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))


days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]


if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    days[1] = 29

day += 1

if day > days[month - 1]:
    day = 1
    month += 1

if month > 12:
    month = 1
    year += 1

print("Next Date:", day, "/", month, "/", year)
```

## 8. Grade Sheet Program

```python
name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAPID: ")

print("Enter marks of 5 subjects:")

pds = int(input("PDS: "))
python_marks = int(input("Python: "))
chem = int(input("Chemistry: "))
eng = int(input("English: "))
phy = int(input("Physics: "))

total = pds + python_marks + chem + eng + phy
percentage = total / 5
cgpa = percentage / 10


if 0 <= cgpa <= 3.4:
    grade = "F"
elif 3.5 <= cgpa <= 5.0:
    grade = "C+"
elif 5.1 <= cgpa <= 6.0:
    grade = "B"
elif 6.1 <= cgpa <= 7.0:
    grade = "B+"
elif 7.1 <= cgpa <= 8.0:
    grade = "A"
elif 8.1 <= cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O"

print("\n----- Grade Sheet -----")
print("Name:", name)
print("Roll Number:", roll)
print("SAPID:", sapid)

print("\nMarks:")
print("PDS:", pds)
print("Python:", python_marks)
print("Chemistry:", chem)
print("English:", eng)
print("Physics:", phy)

print("\nPercentage:", percentage, "%")
print("CGPA:", round(cgpa, 2))
print("Grade:", grade)
```
