```python
class Student:
    school_name = "UPES"   # class variable

    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Aman", 20)
s2 = Student("Riya", 21)

print("School:", s1.school_name)
print("Student 1:", s1.name, s1.age)
print("Student 2:", s2.name, s2.age)
```

### Q2

```python
class Employee:
    company = "ABC Ltd"

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
    def show_details(self):
        print(self.name, self.salary, Employee.company)

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name
e1 = Employee("Aman",50000)
e2 = Employee("Riya",60000)

e1.show_details()
e2.show_details()

Employee.change_company("XYZ ltd")

e1.show_details()
e2.show_details()
```

```python
class MathHelper:
    @staticmethod
    def is_even(num):
        return num % 2 == 0
print(MathHelper.is_even(8))
print(MathHelper.is_even(11))
```

## Q4:

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance

    def deposit(self, amt):
        self.__balance += amt

    def withdraw(self, amt):
        if amt <= self.__balance:
            self.__balance -= amt
        else:
            print("Insufficient balance")

    def show_balance(self):
        print(self.__balance)


acc = BankAccount("Aman", 1000)

acc.deposit(500)
acc.show_balance()

acc.withdraw(300)
acc.show_balance()

acc.withdraw(2000)
```
