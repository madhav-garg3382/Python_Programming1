```python
def top_k(items, k):
    return sorted(items, key=lambda x: x[1], reverse=True)[:k]

print(top_k(items,2))
```

### Question 2

```python
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
```

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] =+ 2
    return d
print(count_by_category(pairs))
```

### Question3
Missing Number (0 to n)

```python
nums = [3,0,1]
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n+1) / 2
    return int(total - sum(nums))
print(missing_number(nums))
```

### Question 4

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    if x not in seen:
        seen.add(x)
        return seen     
   

add_unique(1)
add_unique(2)
```

```python
stream = [1,2,3,4,5]
w = 3
```

```python
def moving_average(stream, w):
    result = []
    window_sum = 0

    for i in range(len(stream)):
        window_sum += stream[i]

        if i >= w:
            window_sum -= stream[i - w]

        avg = window_sum / min(i + 1, w)
        result.append(avg)

    return result


stream = [1, 2, 3, 4, 5]
w = 3
print(moving_average(stream, w))
```
