#### In A list of numbers find prime numbers and replace them with 0


```python
l=["E","P","E","P"]
```

```python
for indx , item in enumerate(l):
    print(indx , item)
    if item == "P":
        l[indx] = 0
```

```python
l
```

```python
l = ["1","2","3","4","5"]
```

```python
for indx, item in enumerate(l):
    print(indx , item)
    if item == l[indx//2]:
        l[indx] = "0"
```

```python
l
```

```python
m=["1","2","3","4","5"]
for indx , item in enumerate(m):
    print(indx , item)
    if int(item) % 2 == 0:
        m[indx] = "0"
```

```python
m
```

```python
grade = [20,30,40,50,60,70,80,90,100]
names = ["ali","ahmed","sara","lina","mona","huda","nora","salma","dina"]
for indx , item in enumerate(grade):
    print(indx , item)
    if item >= 60:
        names[indx] = "pass"

    else:
        names[indx] = "fail"
```
