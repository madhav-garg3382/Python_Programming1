## 6.Hypotenuse using Pythagoras theorem

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = (a*a + b*b) ** 0.5
print("Hypotenuse:", c)
```

## 7.Simple Interest

```python
p = float(input("Enter principal: "))
r = float(input("Enter rate: "))
t = float(input("Enter time: "))

si = (p * r * t) / 100
print("Simple Interest:", si)
```

## 8.Area of triangle (sides given – Heron’s formula)

```python
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

s = (a + b + c) / 2
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
print("Area of triangle:", area)
```

## 9.Convert seconds to hours, minutes, seconds

```python
seconds = int(input("Enter seconds: "))

hours = seconds // 3600
seconds = seconds % 3600
minutes = seconds // 60
seconds = seconds % 60

print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", seconds)
```

## 10.Swap two numbers without extra variable

```python
a = int(input("Enter a: "))
b = int(input("Enter b: "))

a = a + b
b = a - b
a = a - b

print("After swapping:")
print("a =", a)
print("b =", b)
```

## 11.Sum of first n natural numbers

```python
n = int(input("Enter n: "))
sum = n * (n + 1) // 2
print("Sum:", sum)
```

## 12.Truth table for bitwise operators

```python
print("A B  A&B  A|B  A^B")
for a in [0, 1]:
    for b in [0, 1]:
        print(a, b, a & b, a | b, a ^ b)
```

## 13.Left shift and Right shift
