### 1.	Write a Python function to find the maximum and minimum numbers from a sequence of numbers.  

```python
def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]
    
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num
    
    return maximum, minimum

nums = list(map(int, input("Enter numbers: ").split()))
result = find_max_min(nums)
print("Maximum:", result[0])
print("Minimum:", result[1])
```

### 2.	Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.  

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total

n = int(input("Enter number: "))
print("Sum of cubes:", sum_of_cubes(n))
```

### 3.	Write a Python function to print 1 to n using recursion.

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

n = int(input("Enter number: "))
print_numbers(n)
```

### 4.	Write a recursive function to print Fibonacci series upto n terms. 

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

n = int(input("Enter number of terms: "))

for i in range(n):
    print(fibonacci(i), end=" ")
```

### 5.	Write a lambda function to find volume of cone. 

```python
volume = lambda r, h: (1/3) * 3.14 * r * r * h

r = float(input("Enter radius: "))
h = float(input("Enter height: "))

print("Volume of cone:", volume(r, h))
```

### 6.	Write a lambda function which gives tuple of max and min from a list. 

```python
nums = list(map(int, input("Enter numbers: ").split()))

result = (lambda lst: (max(lst), min(lst)))(nums)

print(result)
```
